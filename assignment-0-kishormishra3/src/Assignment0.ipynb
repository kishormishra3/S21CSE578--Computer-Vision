```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
img=[]
for i in range(1,5):
  img.append(cv2.imread(str(i)+'.jpeg',1))
```

```
fig=plt.figure(figsize=(10,10))
columns = 2
rows = 2
j=0
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(img[j])
    plt.axis('off')
    j+=1
plt.show()
```

```
new_img=[]
for i in range(0,4):
  new_img.append(img[i][...,::-1].copy())
```

```
fig=plt.figure(figsize=(10,10))
columns = 2
rows = 2
j=0
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(new_img[j])
    plt.axis('off')
    j+=1
plt.show()
```

```
new= new_img[0]
gray_scale=np.zeros([new.shape[0],new.shape[1]])
for i in range(new.shape[0]):
  for j in range(new.shape[1]):
    gray_scale[i][j]=(new[i][j][0]*0.1)+(new[i][j][1]*0.59)+(new[i][j][2]*0.11)
```

```
plt.imshow(gray_scale)
plt.show()
plt.imsave('images/grey',gray_scale,format='jpeg')
```
